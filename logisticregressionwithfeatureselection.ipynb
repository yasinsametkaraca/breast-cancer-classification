```python
import pandas as pd
import numpy as np
from sklearn.metrics import confusion_matrix,accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from mlxtend.feature_selection import SequentialFeatureSelector
from mlxtend.plotting import plot_sequential_feature_selection as plot_sfs
from sklearn.preprocessing import LabelEncoder,StandardScaler
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score,roc_auc_score,f1_score,precision_score
```

```python
csvPath = r"C:\Users\YSK\Desktop\1030520813_patternrecognition_final_project\breast-cancer.csv"
dataset = pd.read_csv(csvPath)
dataset.drop(['id'], axis = 1, inplace = True)
dataset.isna().sum().sum()
dataset.shape
```

```python
dataset['diagnosis'] = np.where(dataset['diagnosis'] == 'M', 0, 1)
X = dataset.drop(['diagnosis'], axis = 1)
y = dataset['diagnosis']
X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size = 0.2, random_state = 44)
print(X_train.shape, X_valid.shape)
```

```python
clf = LogisticRegression(random_state = 42, max_iter = 150)
sfs = SequentialFeatureSelector(clf,
           k_features=20,
           forward=True,
           floating=False,
           verbose=2,
           scoring='accuracy',
           cv=10)
sfs = sfs.fit(X_train, y_train)
feat_cols = list(sfs.k_feature_idx_)
print(feat_cols)
```

```python
columns=X_train[['area_mean', 'compactness_mean' , 'concavity_mean' , 'perimeter_se', 'compactness_worst','concavity_worst','concave points_worst']]
X_train=X_train.drop(columns, axis=1)
columns=X_valid[['area_mean', 'compactness_mean' , 'concavity_mean' , 'perimeter_se', 'compactness_worst','concavity_worst','concave points_worst']]
X_valid=X_valid.drop(columns, axis=1)
clf = LogisticRegression(random_state = 42, max_iter = 150)
clf.fit(X_train, y_train)
y_predict= clf.predict(X_valid)
```

```python
cMatrix = confusion_matrix(y_valid, y_predict)
print("Confusion Matrix:\n", cMatrix)
print("Classification Report:\n", classification_report(y_valid, y_predict))
print("Accuracy:",accuracy_score(y_valid, y_predict))
print('Sensitivity : ', cMatrix[0, 0]/(cMatrix[0, 0]+cMatrix[0, 1]))
print('Specificity : ', cMatrix[1, 1]/(cMatrix[1, 0]+cMatrix[1, 1]))
print('AUC :',roc_auc_score(y_valid, y_predict))
print('f1-score:', f1_score(y_valid, y_predict))
print('Precision: %.3f' % precision_score(y_valid, y_predict))
```
